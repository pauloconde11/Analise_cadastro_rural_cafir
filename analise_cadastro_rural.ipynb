```python
import pandas as pd
import requests

#documento pode ser baixado em: https://basedosdados.org/dataset/37807a93-219d-45d6-b2ae-5f3f2746a035?table=3a71c300-11ce-404f-8c79-242a2c855f94

file_path = 'bq-results-20260420-020616-1776651080981.csv'

df = pd.read_csv(file_path)
```

```python
df
```

```python
# garantir tipo numérico
df['id_municipio'] = pd.to_numeric(df['id_municipio'], errors='coerce')

# remover nulos
df = df.dropna(subset=['id_municipio'])

# ordenar por data (mais recente primeiro)
df_sorted = df.sort_values(by='data_referencia', ascending=False)

# remover duplicatas (pegar registro mais recente de cada imóvel)
df_unique = df_sorted.drop_duplicates(subset=['id_imovel_receita_federal'])
```

```python
df_unique.shape
```

```python
df_unique['area'].describe()
```

```python
# A distribuição é altamente desigual
```

```python
#top 10 
top10 = df_unique.sort_values(by='area', ascending=False).head(10)
```

```python
top10
```

```python
#os 10 maiores imóveis representam X% da área total

valor_bruto = top10['area'].sum() / df_unique['area'].sum()
porcentagem = round(valor_bruto * 100)
```

```python
porcentagem
```

```python
#os 10 maiores imóveis representam 4% da área total
```

```python
#top 1%

top_1_percent = df_unique.sort_values(by='area', ascending=False).head(int(len(df_unique)*0.01))
valor_bruto_top1 = top_1_percent['area'].sum() / df_unique['area'].sum()
porcentagem_top_1 = round(valor_bruto_top1 *100)
```

```python
porcentagem_top_1
```

```python
#1% dos imoveis representam 37% da área total
```

```python
#top 5%

top_5_percent = df_unique.sort_values(by='area', ascending=False).head(int(len(df_unique)*0.05))
valor_bruto_top5 = top_5_percent['area'].sum() / df_unique['area'].sum()
porcentagem_top_5 = round(valor_bruto_top5 *100)
```

```python
porcentagem_top_5
```

```python
#5% dos imoveis representam 37% da área total
```
